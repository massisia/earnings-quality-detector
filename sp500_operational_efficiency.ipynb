```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define the companies we want to analyze 
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META', 'TSLA', 'JPM', 'JNJ', 'WMT', 'XOM']
```

```python
# Download financial data for all tickers
data = yf.download(tickers, period='1y')
```

```python
companies = {}

for ticker in tickers:
    stock = yf.Ticker(ticker)
    companies[ticker] = stock.info
```

```python
# Double check if code worked: load data for AAPL
companies['AAPL']
```

```python
metrics = []

for ticker in tickers: 
    info = companies[ticker]
    metrics.append({
        'ticker': ticker,
        'company': info.get('shortName'),
        'sector': info.get('sector'),
        'employees': info.get('fullTimeEmployees'),
        'revenue': info.get('totalRevenue'),
        'profit_margin': info.get('profitMargins'),
        'operating_margin': info.get('operatingMargins'),
        'gross_margin': info.get('grossMargins'),
        'return_on_assets': info.get('returnOnAssets'),
        'revenue_per_share': info.get('revenuePerShare')
    })
```

```python
df = pd.DataFrame(metrics) 
df
```
