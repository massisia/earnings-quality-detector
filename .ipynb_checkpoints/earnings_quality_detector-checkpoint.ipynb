```python
# Install required packages
import subprocess
subprocess.run(['py', '-m', 'pip', 'install', 'scikit-learn', 'anthropic', 'python-dotenv'], 
               capture_output=True)

# Core libraries
import pandas as pd
import yfinance as yf
import time
import json
import re
import os
import warnings
from sklearn.preprocessing import MinMaxScaler
from dotenv import load_dotenv
import anthropic

warnings.filterwarnings('ignore')
load_dotenv()
```

```python
# get S&P 500 company list from Github 
url = "https://raw.githubusercontent.com/datasets/s-and-p-500-companies/main/data/constituents.csv"
df_sp500 = pd.read_csv(url) 

print(df_sp500.shape) 
print(df_sp500.columns.tolist())
print(df_sp500.head())
```

```python
df_sp500 = df_sp500[['Symbol', 'Security', 'GICS Sector']]
tickers = df_sp500['Symbol'].tolist()

print(f"Total companies: {len(tickers)}")
print(tickers[:10])
```

```python
def calculate_signals(ticker):
    try: 
        stock = yf.Ticker(ticker) 

        income = stock.financials
        cashflow = stock.cashflow
        balancesheet = stock.balance_sheet

        net_income = income.loc['Net Income'].iloc[0]
        operating_cf = cashflow.loc['Operating Cash Flow'].iloc[0]
        investing_cf = cashflow.loc['Investing Cash Flow'].iloc[0]
        total_assets = balancesheet.loc['Total Assets'].iloc[0]

        # Signal 1: Cash Conversion Ratio
        ccr = operating_cf / net_income 

        # Signal 2: Accrual Ratio 
        accrual = (net_income - operating_cf - investing_cf) / total_assets 

        return {
            'ticker': ticker, 
            'net_income': net_income, 
            'operating_cf': operating_cf,
            'investing_cf': investing_cf,
            'total_assets': total_assets,
            'cash_conversion_ratio': ccr,
            'accrual_ratio': accrual
        }
    
    except Exception as e:
        return None
```

```python
# Pull data for all S&P 500 companies 
results = []

for i, ticker in enumerate(tickers):
    result = calculate_signals(ticker)
    if result:
        results.append(result)
    
    # Progress update every 50 companies
    if (i + 1) % 50 == 0:
        print(f"Processed {i + 1} of {len(tickers)} companies")
    
    # Pause to avoid rate limiting
    time.sleep(0.5)

print(f"Successfully pulled data for {len(results)} companies")
```

```python
# Save raw results to CSV so we don't have to re-pull from yfinance every time
df_raw = pd.DataFrame(results)
df_raw.to_csv('sp500_raw_data.csv', index=False)
print("Raw data saved to sp500_raw_data.csv")

# Load from CSV if it exists, otherwise use results from memory
import os

if os.path.exists('sp500_raw_data.csv'):
    df = pd.read_csv('sp500_raw_data.csv')
    print(f"Loaded {len(df)} companies from cache")
else:
    df = pd.DataFrame(results)
    print(f"Built DataFrame from {len(results)} results")

print(df.shape)
df.head()
```

```python
df = df.merge(df_sp500[['Symbol', 'Security', 'GICS Sector']], 
              left_on='ticker', 
              right_on='Symbol')

df = df.drop(columns=['Symbol'])
print(df.shape)
df.head()
```

```python
scaler = MinMaxScaler() 

# Filter out companies with negative net income. These distort the CCR calculation
df_positive = df[df['net_income'] > 0].copy()
df_negative = df[df['net_income'] <= 0].copy()

print(f"Companies removed: {len(df_negative)}")
print(f"Companies remaining: {len(df_positive)}")
```

```python
# --- Profitable companies: score on both CCR and accrual ratio ---
df_positive['ccr_scaled'] = scaler.fit_transform(df_positive[['cash_conversion_ratio']])
df_positive['accrual_scaled'] = 1 - scaler.fit_transform(df_positive[['accrual_ratio']])
df_positive['earnings_quality_score'] = (df_positive['ccr_scaled'] + df_positive['accrual_scaled']) / 2
df_positive['scoring_method'] = 'Full Score (CCR + Accrual)'

# --- Loss-making companies: score on accrual ratio only ---
df_negative['accrual_scaled'] = 1 - scaler.fit_transform(df_negative[['accrual_ratio']])
df_negative['earnings_quality_score'] = df_negative['accrual_scaled']
df_negative['scoring_method'] = 'Partial Score (Accrual Only)'

# --- Combine back together ---
df_final = pd.concat([df_positive, df_negative]).sort_values('earnings_quality_score', ascending=False).reset_index(drop=True)

print(df_final.shape)
```

```python
df_final[['ticker', 'Security', 'GICS Sector', 'earnings_quality_score', 'scoring_method']].head(10)
```

```python
loss_tickers = df_negative['ticker'].tolist()
print(loss_tickers)
```

```python
# Pull cash & equivalents and multi-year cash flow for loss-making companies
health_results = []

for ticker in loss_tickers:
    try:
        stock = yf.Ticker(ticker)
        
        # Get balance sheet for cash position
        balance = stock.balance_sheet
        cashflow = stock.cashflow
        
        # Cash & equivalents (most recent)
        try:
            cash = balance.loc['Cash And Cash Equivalents'].iloc[0]
        except:
            try:
                cash = balance.loc['Cash Cash Equivalents And Short Term Investments'].iloc[0]
            except:
                cash = balance.loc['Cash Financial'].iloc[0]
        
        # Operating cash flow for last 3 years
        ocf_years = cashflow.loc['Operating Cash Flow'].values[:3]
        
        # Cash burn rate (most recent operating cash flow / cash)
        burn_rate = ocf_years[0] / cash
        
        # Cash runway in quarters (cash / quarterly burn)
        quarterly_burn = abs(ocf_years[0]) / 4
        runway = cash / quarterly_burn if ocf_years[0] < 0 else None
        
        # Cash flow trend - both YoY and 3-year direction
        yoy = ocf_years[0] > ocf_years[1]
        three_year = ocf_years[0] > ocf_years[-1]

        if yoy and three_year:
            trend = 'Consistently Improving'
        elif yoy and not three_year:
            trend = 'Recently Improving'
        elif not yoy and three_year:
            trend = 'Recently Deteriorating'
        else:
            trend = 'Consistently Deteriorating'
                
        health_results.append({
            'ticker': ticker,
            'cash': cash,
            'ocf_current': ocf_years[0],
            'burn_rate': burn_rate,
            'runway_quarters': runway,
            'cash_flow_trend': trend
        })
        
        time.sleep(0.3)
        
    except Exception as e:
        print(f"Error fetching {ticker}: {e}")

print(f"Successfully pulled: {len(health_results)} companies")
```

```python
# Convert to DataFrame and merge sector info
df_health = pd.DataFrame(health_results)
df_health = df_health.merge(df_sp500[['Symbol', 'Security', 'GICS Sector']], 
                             left_on='ticker', 
                             right_on='Symbol').drop(columns=['Symbol'])

df_health['runway_quarters'] = df_health['runway_quarters'].fillna('N/A - Positive OCF')

print(df_health.shape)
df_health.head()
```

```python
# Export our data to JSON for the dashboard
import json

# Prepare profitable companies data
dashboard_data = df_positive[['ticker', 'Security', 'GICS Sector', 
                               'earnings_quality_score', 
                               'cash_conversion_ratio',
                               'accrual_ratio',
                               'net_income',
                               'operating_cf',
                               'scoring_method']].copy()

dashboard_data = dashboard_data.sort_values('earnings_quality_score', ascending=False)
dashboard_data['rank'] = range(1, len(dashboard_data) + 1)

# Export to JSON
dashboard_data.to_json('earnings_quality_data.json', orient='records', indent=2)

# Export loss-making companies
df_health.to_json('health_monitor_data.json', orient='records', indent=2)

print("Data exported successfully")
print(f"Profitable companies: {len(dashboard_data)}")
print(f"Loss-making companies: {len(df_health)}")
```

```python
client = anthropic.Anthropic(api_key=os.getenv("ANTHROPIC_API_KEY"))

def generate_commentary_with_news(company_data):
    ticker = company_data['ticker']
    company = company_data['Security']
    score = company_data['earnings_quality_score']
    ccr = company_data['cash_conversion_ratio']
    accrual = company_data['accrual_ratio']
    net_income = company_data['net_income']
    operating_cf = company_data['operating_cf']
    
    ni_str = f"${net_income/1e9:.2f}B" if abs(net_income) >= 1e9 else f"${net_income/1e6:.0f}M"
    ocf_str = f"${operating_cf/1e9:.2f}B" if abs(operating_cf) >= 1e9 else f"${operating_cf/1e6:.0f}M"
    
    message = client.messages.create(
        model="claude-sonnet-4-6",
        max_tokens=1024,
        temperature=0,
        tools=[{
            "type": "web_search_20250305",
            "name": "web_search"
        }],
        messages=[
            {
                "role": "user",
                "content": f"""Search for recent news about {company} ({ticker}) focusing on earnings, cash flow, and financial performance.

Then evaluate this company's earnings quality from the perspective of 5 different analysts:

1. THE BULL: Finds the strongest case for high earnings quality
2. THE BEAR: Finds every red flag and reason to be skeptical
3. THE FORENSIC ACCOUNTANT: Focuses purely on accounting signals and what the numbers reveal
4. THE MOMENTUM TRADER: Focuses on whether recent trends support or undermine the score
5. THE CREDIT ANALYST: Focuses on cash generation sustainability and balance sheet implications

Financial metrics:
- Earnings Quality Score: {score:.3f} (scale 0-1, higher is better)
- Cash Conversion Ratio: {ccr:.3f} (above 1.0 is strong, measures OCF/Net Income)
- Accrual Ratio: {accrual:.4f} (negative is better)
- Net Income: {ni_str}
- Operating Cash Flow: {ocf_str}

After all 5 perspectives, write a SYNTHESIS of 3 sentences that captures the most important insight from the council. Reference both the numbers and the news. Be direct, no filler.

Format your response as:
BULL: [one sentence]
BEAR: [one sentence]
FORENSIC: [one sentence]
MOMENTUM: [one sentence]
CREDIT: [one sentence]
SYNTHESIS: [three sentences]"""
            }
        ]
    )
    
    full_response = ' '.join([block.text for block in message.content if hasattr(block, 'text')])
    return full_response

# Test on top company first
top_10 = df_positive.nlargest(10, 'earnings_quality_score')
top_company = top_10.iloc[0]
commentary = generate_commentary_with_news(top_company)
print(commentary)
```

```python
commentaries = {}

# Already generated for top company (CSGP) as a test
commentaries['CSGP'] = """[Here is the full five-analyst council evaluation, grounded in CoStar's actual reported financials:

---

**BULL:**  The $430M in operating cash flow — against only $7M in GAAP net income — combined with Adjusted Net Income of $364M (up 23%) and Adjusted EBITDA of $442M (up 83%) reveals a business generating real, compounding cash that GAAP accounting is temporarily obscuring due to acquisition costs. 

**BEAR:**  Net profit margins have collapsed from 5.1% to 0.2%, earnings have declined 17% per year over five years, and diluted EPS cratered 94% year-over-year — a pattern that makes the 0.586 earnings quality score look generous, not conservative. 

**FORENSIC:**  The cash flow statement tells the real story: $194M in stock-based compensation, $263M in D&A, and $141M in deferred commission amortization are the primary bridges between $7M net income and $430M OCF  — meaning the Cash Conversion Ratio of 61.4x is almost entirely a function of non-cash add-backs and acquisition accounting, not organic cash generation, and the accrual ratio of 0.2271 (positive, not negative) signals earnings are *lagging* cash, not leading it.

**MOMENTUM:**  Revenue acceleration is undeniable — Q1 2026 grew 23% YoY to $897M, following Q3 2025's 20% growth to $834M  — but  quarterly EPS has swung between a loss of $0.07 and a high of $0.15 over the past six quarters, leaving investors focused on how much of that growing revenue is translating into lasting earnings power. 

**CREDIT:**  Cash, cash equivalents, and restricted cash declined sharply from $4.68B to $1.73B in 2025, as the company generated $430M in operating cash flow but deployed $2.82B in investing activities for acquisitions  — and  with a $500M buyback completed in Q4 2025 and a $700M repurchase planned for 2026, the balance sheet is being drawn down aggressively even as free cash flow of only $123M provides a thin margin of safety. 

---

**SYNTHESIS:** CoStar's $430M OCF against $7M net income is not a sign of earnings quality — it is a sign of a company in heavy investment mode, where acquisition costs (Matterport, Domain), $194M in stock-based compensation, and $263M in D&A are masking the true economic cost of growth.  The company has delivered 57 consecutive quarters of double-digit revenue growth, with Q3 2025 reaching $834M at 20% YoY , proving the top-line engine is real — but  most of that $3.2B in revenue is not converting to profit today, and with trailing earnings declining 17% per year over five years, the profit side has not yet matched the revenue story.  The 0.586 earnings quality score is fair: this is a high-quality *business* with a low-quality *earnings line*, and the investment thesis lives or dies on whether the Homes.com and residential bets eventually close the gap between $442M Adjusted EBITDA and $7M GAAP net income.]"""

for i, row in top_10.iterrows():
    if i == 0:  # Skip CSGP
        continue
    ticker = row['ticker']
    print(f"Generating commentary for {ticker} ({i+1}/10)...")
    try:
        commentary = generate_commentary_with_news(row)
        commentaries[ticker] = commentary
        time.sleep(2)
    except Exception as e:
        print(f"Error on {ticker}: {e}")
        commentaries[ticker] = "Commentary unavailable."

print("Done.")
```

```python
import json

with open('commentaries.json', 'w', encoding='utf-8') as f:
    json.dump(commentaries, f, indent=2)

print(f"Saved {len(commentaries)} commentaries.")
print(list(commentaries.keys()))
```

```python
import json
import re

# Load the raw commentaries
with open('commentaries.json', 'r', encoding='utf-8') as f:
    raw = json.load(f)

# Extract just the synthesis from each commentary and clean it
clean = {}
for ticker, text in raw.items():
    # Extract synthesis section
    match = re.search(r'SYNTHESIS[:\*]*\s*([\s\S]*?)(?:\n\n---|\Z)', text, re.IGNORECASE)
    if match:
        synthesis = match.group(1).strip()
    else:
        synthesis = text.strip()
    
    # Clean problematic characters
    synthesis = synthesis.replace('`', "'")
    synthesis = synthesis.replace('\\', '')
    synthesis = synthesis.replace('"', "'")
    synthesis = re.sub(r'\*+', '', synthesis)  # remove markdown bold
    synthesis = re.sub(r'\n+', ' ', synthesis)  # flatten newlines
    synthesis = synthesis.strip()
    
    clean[ticker] = synthesis

# Save clean version
with open('commentaries.json', 'w', encoding='utf-8') as f:
    json.dump(clean, f, indent=2, ensure_ascii=False)

print("Cleaned commentaries saved.")
for ticker, text in clean.items():
    print(f"\n{ticker}: {text[:100]}...")
```

```python
dashboard_html = open('earnings_quality_data.json').read()
health_html = open('health_monitor_data.json').read()
commentary_html = open('commentaries.json').read()

html = f"""<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8">
<meta name="viewport" content="width=device-width, initial-scale=1.0">
<title>S&P 500 Earnings Quality Detector</title>
<style>
  * {{ margin: 0; padding: 0; box-sizing: border-box; }}
  body {{ font-family: 'Segoe UI', system-ui, sans-serif; background: #0f1117; color: #e8eaf0; min-height: 100vh; }}
  header {{ padding: 40px 60px 30px; border-bottom: 1px solid #1e2130; }}
  header h1 {{ font-size: 28px; font-weight: 700; color: #fff; letter-spacing: -0.5px; }}
  header p {{ margin-top: 6px; font-size: 14px; color: #6b7280; }}
  .tabs {{ display: flex; gap: 4px; padding: 20px 60px 0; border-bottom: 1px solid #1e2130; }}
  .tab {{ padding: 10px 24px; font-size: 13px; font-weight: 500; color: #6b7280; cursor: pointer; border-bottom: 2px solid transparent; transition: all 0.2s; }}
  .tab.active {{ color: #4f9cf9; border-bottom: 2px solid #4f9cf9; }}
  .tab:hover {{ color: #e8eaf0; }}
  .content {{ padding: 30px 60px; }}
  .stats-row {{ display: grid; grid-template-columns: repeat(3, 1fr); gap: 16px; margin-bottom: 30px; max-width: 700px; margin-left: auto; margin-right: auto; }}
  .stat-card {{ background: #161b27; border: 1px solid #1e2130; border-radius: 10px; padding: 20px 24px; }}
  .stat-card .label {{ font-size: 11px; text-transform: uppercase; letter-spacing: 1px; color: #6b7280; margin-bottom: 8px; }}
  .stat-card .value {{ font-size: 28px; font-weight: 700; color: #fff; }}
  .stat-card .sub {{ font-size: 12px; color: #6b7280; margin-top: 4px; }}
  .panel {{ background: #161b27; border: 1px solid #1e2130; border-radius: 10px; padding: 24px; margin-bottom: 20px; }}
  .panel h2 {{ font-size: 15px; font-weight: 600; color: #fff; margin-bottom: 20px; }}
  .two-col {{ display: grid; grid-template-columns: 1fr 1fr; gap: 20px; margin-bottom: 20px; }}
  .leaderboard {{ display: flex; flex-direction: column; gap: 8px; }}
  .lb-row {{ display: flex; align-items: center; gap: 12px; padding: 10px 14px; background: #0f1117; border-radius: 8px; border: 1px solid #1e2130; transition: background 0.15s; }}
  .lb-row:hover {{ background: #1a1f2e; }}
  .lb-rank {{ font-size: 12px; font-weight: 700; color: #6b7280; width: 28px; text-align: center; }}
  .lb-rank.gold {{ color: #f59e0b; }}
  .lb-rank.silver {{ color: #9ca3af; }}
  .lb-rank.bronze {{ color: #b45309; }}
  .lb-ticker {{ font-family: monospace; font-size: 13px; font-weight: 700; color: #4f9cf9; width: 52px; }}
  .lb-name {{ flex: 1; font-size: 13px; color: #d1d5db; white-space: nowrap; overflow: hidden; text-overflow: ellipsis; }}
  .lb-sector {{ font-size: 11px; color: #6b7280; width: 140px; white-space: nowrap; overflow: hidden; text-overflow: ellipsis; }}
  .lb-score-wrap {{ display: flex; align-items: center; gap: 8px; width: 130px; }}
  .lb-bar-track {{ flex: 1; height: 5px; background: #1e2130; border-radius: 3px; overflow: hidden; }}
  .lb-bar-fill {{ height: 100%; border-radius: 3px; }}
  .lb-score-val {{ font-size: 12px; font-weight: 600; width: 36px; text-align: right; }}
  .controls {{ display: flex; gap: 12px; margin-bottom: 24px; align-items: center; flex-wrap: wrap; }}
  input[type="text"] {{ background: #161b27; border: 1px solid #2a3040; border-radius: 6px; padding: 8px 14px; color: #e8eaf0; font-size: 13px; width: 220px; outline: none; }}
  input[type="text"]:focus {{ border-color: #4f9cf9; }}
  select {{ background: #161b27; border: 1px solid #2a3040; border-radius: 6px; padding: 8px 14px; color: #e8eaf0; font-size: 13px; outline: none; cursor: pointer; }}
  table {{ width: 100%; border-collapse: collapse; font-size: 13px; }}
  thead th {{ text-align: left; padding: 10px 14px; font-size: 11px; text-transform: uppercase; letter-spacing: 0.8px; color: #6b7280; border-bottom: 1px solid #1e2130; }}
  tbody tr {{ border-bottom: 1px solid #1a1f2e; transition: background 0.15s; }}
  tbody tr:hover {{ background: #1e2435; }}
  tbody tr.highlighted {{ background: #1a2a3a; border-left: 3px solid #4f9cf9; }}
  tbody td {{ padding: 11px 14px; color: #d1d5db; }}
  .rank {{ color: #6b7280; font-size: 12px; }}
  .ticker {{ font-weight: 700; color: #4f9cf9; font-family: monospace; }}
  .score-bar {{ display: flex; align-items: center; gap: 10px; }}
  .bar-track {{ width: 80px; height: 5px; background: #1e2130; border-radius: 3px; overflow: hidden; }}
  .bar-fill {{ height: 100%; border-radius: 3px; }}
  .score-val {{ font-size: 12px; color: #9ca3af; min-width: 36px; }}
  .sector-badge {{ background: #1e2435; color: #9ca3af; padding: 2px 8px; border-radius: 4px; font-size: 11px; }}
  .pagination {{ display: flex; justify-content: space-between; align-items: center; margin-top: 16px; font-size: 13px; color: #6b7280; }}
  .page-btns {{ display: flex; gap: 6px; align-items: center; }}
  .page-btn {{ padding: 5px 12px; border: 1px solid #2a3040; border-radius: 5px; background: transparent; color: #9ca3af; cursor: pointer; font-size: 12px; transition: all 0.15s; }}
  .page-btn:hover {{ background: #1e2435; color: #fff; }}
  .page-btn.active {{ background: #4f9cf9; border-color: #4f9cf9; color: #fff; }}
  .page-btn:disabled {{ opacity: 0.3; cursor: not-allowed; }}
  .tab-content {{ display: none; }}
  .tab-content.active {{ display: block; }}
  .methodology-box {{ background: #0f1117; border: 1px solid #2a3040; border-radius: 8px; padding: 16px 20px; margin-bottom: 24px; font-size: 13px; color: #9ca3af; line-height: 1.7; }}
  .methodology-box strong {{ color: #d1d5db; }}
  .badge {{ display: inline-block; padding: 3px 10px; border-radius: 20px; font-size: 11px; font-weight: 500; }}
  .trend-ci {{ background: #052e16; color: #4ade80; }}
  .trend-ri {{ background: #1c3626; color: #86efac; }}
  .trend-rd {{ background: #3b1a1a; color: #fca5a5; }}
  .trend-cd {{ background: #2d0f0f; color: #f87171; }}
  .strip-container {{ position: relative; width: 100%; }}
  .strip-row {{ display: flex; align-items: center; margin-bottom: 18px; gap: 12px; }}
  .strip-sector-label {{ width: 180px; font-size: 12px; color: #9ca3af; text-align: right; flex-shrink: 0; }}
  .strip-track {{ flex: 1; height: 24px; position: relative; }}
  .strip-avg-line {{ position: absolute; top: 0; bottom: 0; width: 2px; background: #4f9cf9; opacity: 0.6; transform: translateX(-50%); z-index: 1; }}
  .strip-dot {{ position: absolute; top: 50%; transform: translate(-50%, -50%); width: 8px; height: 8px; border-radius: 50%; cursor: pointer; transition: all 0.15s; z-index: 2; }}
  .strip-dot:hover {{ width: 12px; height: 12px; z-index: 10; }}
  .strip-axis {{ display: flex; margin-left: 192px; margin-top: 4px; justify-content: space-between; font-size: 11px; color: #6b7280; }}
  .strip-avg-label {{ font-size: 11px; color: #4f9cf9; margin-left: 4px; }}
  .tooltip {{ position: fixed; background: #1e2435; border: 1px solid #2a3040; border-radius: 8px; padding: 10px 14px; font-size: 12px; color: #e8eaf0; pointer-events: none; z-index: 1000; display: none; box-shadow: 0 4px 20px rgba(0,0,0,0.5); }}
  .tooltip .tt-ticker {{ font-weight: 700; color: #4f9cf9; font-family: monospace; }}
  .tooltip .tt-score {{ font-weight: 700; margin-top: 2px; }}
</style>
</head>
<body>

<header>
  <h1>S&P 500 Earnings Quality Detector</h1>
  <p>Data sourced from Yahoo Finance via the yfinance API, covering S&P 500 constituents. Of the 503 companies in the index, 498 returned sufficient financial data for analysis; 5 were excluded due to incomplete filings. Of the 498, 469 reported positive net income and received a full earnings quality score based on Cash Conversion Ratio and Accrual Ratio. The remaining 27 reported net losses and are evaluated separately on cash trajectory metrics. Built by Massi.</p>
</header>

<div class="tabs">
  <div class="tab active" onclick="switchTab('overview', this)">Overview</div>
  <div class="tab" onclick="switchTab('rankings', this)">Full Rankings</div>
  <div class="tab" onclick="switchTab('health', this)">Negative Net Income: Cash Analysis</div>
</div>

<div id="overview" class="tab-content active content">
  <div class="stats-row">
    <div class="stat-card">
      <div class="label">Companies Analyzed</div>
      <div class="value">498</div>
      <div class="sub">S&amp;P 500 universe</div>
    </div>
    <div class="stat-card">
      <div class="label">Positive Net Income</div>
      <div class="value">469</div>
      <div class="sub">Full CCR + Accrual score</div>
    </div>
    <div class="stat-card">
      <div class="label">Negative Net Income</div>
      <div class="value">27</div>
      <div class="sub">Scored separately</div>
    </div>
  </div>

  <div class="two-col">
    <div class="panel">
      <h2>Top 15 &#8212; Highest Earnings Quality</h2>
      <div class="leaderboard" id="top-leaderboard"></div>
    </div>
    <div class="panel">
      <h2>Bottom 15 &#8212; Lowest Earnings Quality</h2>
      <div class="leaderboard" id="bottom-leaderboard"></div>
    </div>
  </div>

  <div class="panel">
    <h2>Earnings Quality Distribution by Sector</h2>
    <p style="font-size:12px;color:#6b7280;margin-bottom:20px;">Each dot is a company. Blue line = sector average. Hover to identify. Click to highlight in Full Rankings.</p>
    <div class="strip-container" id="strip-plot"></div>
    <div class="strip-axis">
      <span>0.0</span><span>0.1</span><span>0.2</span><span>0.3</span><span>0.4</span><span>0.5</span><span>0.6+</span>
    </div>
  </div>
</div>

<div id="rankings" class="tab-content content">
    <div class="methodology-box">
        <strong>How the Earnings Quality Score is calculated.</strong><br><br>
        Each company is scored on two signals:<br><br>
        <strong>Cash Conversion Ratio (CCR)</strong>: Operating Cash Flow divided by Net Income. Measures how much of every dollar of reported profit actually became cash. A ratio close to or above 1.0 indicates high earnings quality. A ratio significantly below 1.0 suggests reported profits are not being realized in cash, which can signal aggressive revenue recognition or understated expenses.<br><br>
        <strong>Accrual Ratio</strong>: Net Income minus Operating Cash Flow minus Investing Cash Flow, divided by Total Assets. Measures how much of a company's earnings are driven by non-cash accounting entries rather than actual cash generation. A lower or negative accrual ratio is better; it means the company's cash performance exceeds what accounting rules require it to report.<br><br>
        Both signals are normalized to a 0-1 scale using Min-Max scaling so they contribute equally to the final score. The composite Earnings Quality Score is their simple average. A score closer to 1.0 indicates higher earnings reliability. Only companies with positive net income are included here. Note: the 27 companies reporting net losses are evaluated separately under the Negative Net Income tab.
    </div>
  <div class="controls">
    <input type="text" id="search" placeholder="Search company or ticker..." oninput="filterTable()">
    <select id="sector-filter" onchange="filterTable()">
      <option value="">All Sectors</option>
    </select>
    <select id="sort-select" onchange="sortTable()">
      <option value="rank">Sort: Rank</option>
      <option value="earnings_quality_score">Sort: Score</option>
      <option value="cash_conversion_ratio">Sort: CCR</option>
      <option value="accrual_ratio">Sort: Accrual Ratio</option>
    </select>
    <span id="result-count" style="color:#6b7280;font-size:13px;"></span>
  </div>
  <div style="display:flex;gap:16px;margin-bottom:16px;flex-wrap:wrap;font-size:12px;color:#9ca3af;">
    <span style="color:#6b7280;font-size:11px;text-transform:uppercase;letter-spacing:1px;align-self:center;">Columns:</span>
    <label style="display:flex;align-items:center;gap:6px;cursor:pointer;"><input type="checkbox" checked onchange="toggleCol(3, this)"> Sector</label>
    <label style="display:flex;align-items:center;gap:6px;cursor:pointer;"><input type="checkbox" checked onchange="toggleCol(4, this)"> Quality Score</label>
    <label style="display:flex;align-items:center;gap:6px;cursor:pointer;"><input type="checkbox" checked onchange="toggleCol(5, this)"> Net Income</label>
    <label style="display:flex;align-items:center;gap:6px;cursor:pointer;"><input type="checkbox" checked onchange="toggleCol(6, this)"> Operating CF</label>
    <label style="display:flex;align-items:center;gap:6px;cursor:pointer;"><input type="checkbox" checked onchange="toggleCol(7, this)"> CCR</label>
    <label style="display:flex;align-items:center;gap:6px;cursor:pointer;"><input type="checkbox" checked onchange="toggleCol(8, this)"> Accrual Ratio</label>
  </div>
  <div class="panel" style="padding:0;overflow:hidden;">
 <table id="rankings-table">
      <thead>
        <tr>
          <th>Rank</th><th>Ticker</th><th>Company</th><th>Sector</th>
          <th>Quality Score</th><th>Net Income</th><th>Operating CF</th><th>CCR</th><th>Accrual Ratio</th>
        </tr>
      </thead>
      <tbody id="table-body"></tbody>
    </table>
  </div>
  <div class="pagination">
    <span id="page-info"></span>
    <div class="page-btns">
      <button class="page-btn" id="prev-btn" onclick="changePage(-1)">&#8592; Prev</button>
      <div id="page-numbers"></div>
      <button class="page-btn" id="next-btn" onclick="changePage(1)">Next &#8594;</button>
    </div>
  </div>
</div>

<div id="health" class="tab-content content">
  <div class="methodology-box">
    <strong>Why these companies are scored differently.</strong><br>
    The Cash Conversion Ratio (CCR) becomes unreliable when net income is negative. A company posting high net income but low operating cash flow signals poor earnings quality, but the inverse does not hold when net income itself is negative. For these 27 companies, we evaluate cash trajectory instead.<br><br>
    <strong>Cash Position</strong>: Total cash and cash equivalents on the balance sheet. Indicates how much liquidity the company has available.<br><br>
    <strong>Operating Cash Flow (OCF)</strong>: Actual cash generated or consumed by core business operations. A company can report a net loss but still generate positive OCF. This often occurs because losses are driven by non-cash charges like depreciation or impairment rather than actual cash outflows. Shown in green if positive, red if negative.<br><br>
    <strong>OCF / Cash Ratio</strong>: Operating cash flow relative to cash reserves. A ratio above 1.0 means the company is generating more cash than it currently holds. A negative ratio means cash is being consumed.<br><br>
    <strong>Cash Flow Trend</strong>: Reflects both short-term (year-over-year) and long-term (3-year) direction simultaneously:<br>
    &nbsp;&nbsp;&#8226; <strong>Consistently Improving</strong>: OCF is better than both last year and 3 years ago<br>
    &nbsp;&nbsp;&#8226; <strong>Recently Improving</strong>: OCF is better than last year but worse than 3 years ago; recovery signal<br>
    &nbsp;&nbsp;&#8226; <strong>Recently Deteriorating</strong>: OCF is worse than last year but better than 3 years ago; recent setback on otherwise improving trajectory<br>
    &nbsp;&nbsp;&#8226; <strong>Consistently Deteriorating</strong>: OCF is worse than both last year and 3 years ago; sustained decline
  </div>
 <div style="display:flex;gap:16px;margin-bottom:16px;flex-wrap:wrap;font-size:12px;color:#9ca3af;">
    <span style="color:#6b7280;font-size:11px;text-transform:uppercase;letter-spacing:1px;align-self:center;">Columns:</span>
    <label style="display:flex;align-items:center;gap:6px;cursor:pointer;"><input type="checkbox" checked onchange="toggleHealthCol(2, this)"> Sector</label>
    <label style="display:flex;align-items:center;gap:6px;cursor:pointer;"><input type="checkbox" checked onchange="toggleHealthCol(3, this)"> Cash Position</label>
    <label style="display:flex;align-items:center;gap:6px;cursor:pointer;"><input type="checkbox" checked onchange="toggleHealthCol(4, this)"> Operating Cash Flow</label>
    <label style="display:flex;align-items:center;gap:6px;cursor:pointer;"><input type="checkbox" checked onchange="toggleHealthCol(5, this)"> OCF / Cash Ratio</label>
    <label style="display:flex;align-items:center;gap:6px;cursor:pointer;"><input type="checkbox" checked onchange="toggleHealthCol(6, this)"> Cash Flow Trend</label>
  </div>
  <div class="panel" style="padding:0;overflow:hidden;">
    <table id="health-table">
      <thead>
        <tr>
          <th>Ticker</th><th>Company</th><th>Sector</th>
          <th>Cash Position</th><th>Operating Cash Flow</th>
          <th>OCF / Cash Ratio</th><th>Cash Flow Trend</th>
        </tr>
      </thead>
      <tbody id="health-body"></tbody>
    </table>
  </div>
</div>

<div class="tooltip" id="tooltip">
  <div class="tt-ticker" id="tt-ticker"></div>
  <div id="tt-name" style="color:#d1d5db;margin:2px 0;"></div>
  <div class="tt-score" id="tt-score"></div>
  <div style="font-size:11px;color:#6b7280;margin-top:4px;">Click to highlight in rankings</div>
</div>

<script>
const eqData = {dashboard_html};
const healthData = {health_html};
const commentaries = {commentary_html};

let filteredData = [...eqData];
let currentPage = 1;
const rowsPerPage = 20;
let highlightedTicker = null;

function switchTab(name, el) {{
  document.querySelectorAll('.tab-content').forEach(t => t.classList.remove('active'));
  document.querySelectorAll('.tab').forEach(t => t.classList.remove('active'));
  document.getElementById(name).classList.add('active');
  el.classList.add('active');
  if (name === 'rankings' && highlightedTicker) {{
    setTimeout(() => scrollToHighlighted(), 100);
  }}
}}

function scoreColor(score) {{
  if (score >= 0.5) return '#4ade80';
  if (score >= 0.35) return '#4f9cf9';
  if (score >= 0.2) return '#facc15';
  return '#f87171';
}}

function buildLeaderboard(containerId, data, isBottom) {{
  const container = document.getElementById(containerId);
  data.forEach((d, i) => {{
    const score = d.earnings_quality_score;
    const pct = Math.min(score * 100, 100).toFixed(0);
    const rankLabel = isBottom ? (eqData.length - data.length + i + 1) : (i + 1);
    const rankClass = !isBottom && i === 0 ? 'gold' : !isBottom && i === 1 ? 'silver' : !isBottom && i === 2 ? 'bronze' : '';
    container.innerHTML += `
      <div class="lb-row">
        <div class="lb-rank ${{rankClass}}">#${{rankLabel}}</div>
        <div class="lb-ticker">${{d.ticker}}</div>
        <div class="lb-name">${{d.Security}}</div>
        <div class="lb-sector">${{d['GICS Sector']}}</div>
        <div class="lb-score-wrap">
          <div class="lb-bar-track"><div class="lb-bar-fill" style="width:${{pct}}%;background:${{scoreColor(score)}}"></div></div>
          <span class="lb-score-val" style="color:${{scoreColor(score)}}">${{score.toFixed(3)}}</span>
        </div>
      </div>`;
  }});
}}

function toggleCommentary(row, ticker) {{
  const commentaryRow = document.getElementById('commentary-' + ticker);
  if (!commentaryRow) return;
  const isVisible = commentaryRow.style.display !== 'none';
  commentaryRow.style.display = isVisible ? 'none' : 'table-row';
  row.style.background = isVisible ? '' : '#1a2a3a';
  if (!isVisible && commentaries[ticker]) {{
    const div = commentaryRow.querySelector('div');
    if (div) div.textContent = commentaries[ticker];
  }}
}}

function initLeaderboards() {{
  const top15 = eqData.slice(0, 15);
  const bottom15 = [...eqData].sort((a,b) => a.earnings_quality_score - b.earnings_quality_score).slice(0, 15).reverse();
  buildLeaderboard('top-leaderboard', top15, false);
  buildLeaderboard('bottom-leaderboard', bottom15, true);
}}

function initStripPlot() {{
  const sectors = {{}};
  eqData.forEach(d => {{
    const s = d['GICS Sector'];
    if (!sectors[s]) sectors[s] = [];
    sectors[s].push(d);
  }});

  const sectorAvgs = Object.entries(sectors).map(([name, companies]) => ({{
    name,
    companies,
    avg: companies.reduce((s, d) => s + d.earnings_quality_score, 0) / companies.length
  }})).sort((a,b) => b.avg - a.avg);

  const container = document.getElementById('strip-plot');
  const maxScore = 0.65;
  const tooltip = document.getElementById('tooltip');

  sectorAvgs.forEach(({{ name, companies, avg }}) => {{
    const row = document.createElement('div');
    row.className = 'strip-row';

    const label = document.createElement('div');
    label.className = 'strip-sector-label';
    label.innerHTML = `${{name}}<br><span class="strip-avg-label">avg ${{avg.toFixed(3)}}</span>`;

    const track = document.createElement('div');
    track.className = 'strip-track';

    const avgLine = document.createElement('div');
    avgLine.className = 'strip-avg-line';
    avgLine.style.left = (avg / maxScore * 100) + '%';
    track.appendChild(avgLine);

    companies.forEach(d => {{
      const dot = document.createElement('div');
      dot.className = 'strip-dot';
      const pct = Math.min(d.earnings_quality_score / maxScore * 100, 99);
      dot.style.left = pct + '%';
      dot.style.background = scoreColor(d.earnings_quality_score);
      dot.style.opacity = '0.75';

      dot.addEventListener('mouseenter', (e) => {{
        tooltip.style.display = 'block';
        document.getElementById('tt-ticker').textContent = d.ticker;
        document.getElementById('tt-name').textContent = d.Security;
        document.getElementById('tt-score').textContent = 'Score: ' + d.earnings_quality_score.toFixed(3);
        document.getElementById('tt-score').style.color = scoreColor(d.earnings_quality_score);
      }});

      dot.addEventListener('mousemove', (e) => {{
        tooltip.style.left = (e.clientX + 14) + 'px';
        tooltip.style.top = (e.clientY - 10) + 'px';
      }});

      dot.addEventListener('mouseleave', () => {{ tooltip.style.display = 'none'; }});

      dot.addEventListener('click', () => {{
        highlightedTicker = d.ticker;
        document.querySelectorAll('.tab-content').forEach(t => t.classList.remove('active'));
        document.querySelectorAll('.tab').forEach(t => t.classList.remove('active'));
        document.getElementById('rankings').classList.add('active');
        document.querySelectorAll('.tab')[1].classList.add('active');
        document.getElementById('search').value = d.ticker;
        filterTable();
        setTimeout(() => scrollToHighlighted(), 150);
      }});

      track.appendChild(dot);
    }});

    row.appendChild(label);
    row.appendChild(track);
    container.appendChild(row);
  }});
}}

function scrollToHighlighted() {{
  const highlighted = document.querySelector('tbody tr.highlighted');
  if (highlighted) highlighted.scrollIntoView({{ behavior: 'smooth', block: 'center' }});
}}

function initSectorFilter() {{
  const sectors = [...new Set(eqData.map(d => d['GICS Sector']))].sort();
  const sel = document.getElementById('sector-filter');
  sectors.forEach(s => {{
    const opt = document.createElement('option');
    opt.value = s; opt.textContent = s;
    sel.appendChild(opt);
  }});
}}

function filterTable() {{
  const search = document.getElementById('search').value.toLowerCase();
  const sector = document.getElementById('sector-filter').value;
  filteredData = eqData.filter(d => {{
    const matchSearch = d.ticker.toLowerCase().includes(search) || d.Security.toLowerCase().includes(search);
    const matchSector = !sector || d['GICS Sector'] === sector;
    return matchSearch && matchSector;
  }});
  currentPage = 1;
  renderTable();
}}

function sortTable() {{
  const key = document.getElementById('sort-select').value;
  filteredData.sort((a,b) => key === 'rank' ? a.rank - b.rank : b[key] - a[key]);
  renderTable();
}}

function renderTable() {{
  const start = (currentPage - 1) * rowsPerPage;
  const page = filteredData.slice(start, start + rowsPerPage);
  const tbody = document.getElementById('table-body');
  tbody.innerHTML = '';
  page.forEach(d => {{
    const score = d.earnings_quality_score;
    const pct = Math.min(score * 100, 100).toFixed(0);
    const isHighlighted = d.ticker === highlightedTicker;
    tbody.innerHTML += `
      <tr class="${{isHighlighted ? 'highlighted' : ''}}" style="cursor:pointer;" onclick="toggleCommentary(this, '${{d.ticker}}')">
        <td class="rank">#${{d.rank}}</td>
        <td class="ticker">${{d.ticker}}</td>
        <td>${{d.Security}}</td>
        <td><span class="sector-badge">${{d['GICS Sector']}}</span></td>
        <td>
          <div class="score-bar">
            <div class="bar-track"><div class="bar-fill" style="width:${{pct}}%;background:${{scoreColor(score)}}"></div></div>
            <span class="score-val">${{score.toFixed(3)}}</span>
          </div>
        </td>
       <td>${{Math.abs(d.net_income) >= 1e9 ? (d.net_income / 1e9).toFixed(2) + 'B' : (d.net_income / 1e6).toFixed(0) + 'M'}}</td>
        <td>${{Math.abs(d.operating_cf) >= 1e9 ? (d.operating_cf / 1e9).toFixed(2) + 'B' : (d.operating_cf / 1e6).toFixed(0) + 'M'}}</td>
        <td>${{d.cash_conversion_ratio.toFixed(3)}}</td>
        <td>${{d.accrual_ratio.toFixed(4)}}</td>
      </tr>
      <tr id="commentary-${{d.ticker}}" style="display:none;">
        <td colspan="9" style="padding:0;">
          <div style="padding:16px 20px;background:#0d1420;border-left:3px solid #4f9cf9;font-size:13px;color:#9ca3af;line-height:1.8;white-space:pre-wrap;">${{commentaries['${{d.ticker}}'] || 'No AI commentary available for this company.'}}</div>
        </td>
      </tr>`;
  }});
  document.getElementById('result-count').textContent = `${{filteredData.length}} companies`;
  document.getElementById('page-info').textContent = `Page ${{currentPage}} of ${{Math.ceil(filteredData.length/rowsPerPage)}}`;
  renderPageNumbers();
}}

function renderPageNumbers() {{
  const total = Math.ceil(filteredData.length / rowsPerPage);
  const container = document.getElementById('page-numbers');
  container.innerHTML = '';
  for (let i = 1; i <= total; i++) {{
    if (i === 1 || i === total || Math.abs(i - currentPage) <= 1) {{
      const btn = document.createElement('button');
      btn.className = 'page-btn' + (i === currentPage ? ' active' : '');
      btn.textContent = i;
      btn.onclick = () => {{ currentPage = i; renderTable(); }};
      container.appendChild(btn);
    }} else if (Math.abs(i - currentPage) === 2) {{
      const span = document.createElement('span');
      span.textContent = '...';
      span.style.cssText = 'color:#6b7280;padding:0 4px';
      container.appendChild(span);
    }}
  }}
  document.getElementById('prev-btn').disabled = currentPage === 1;
  document.getElementById('next-btn').disabled = currentPage === total;
}}

function changePage(dir) {{
  const total = Math.ceil(filteredData.length / rowsPerPage);
  currentPage = Math.max(1, Math.min(total, currentPage + dir));
  renderTable();
}}

function toggleCol(index, checkbox) {{
  const table = document.getElementById('rankings-table') || document.querySelector('#rankings table');
  const rows = table ? table.querySelectorAll('tr') : [];
  rows.forEach(row => {{
    const cells = row.querySelectorAll('th, td');
    if (cells[index]) cells[index].style.display = checkbox.checked ? '' : 'none';
  }});
}}

function toggleHealthCol(index, checkbox) {{
  const table = document.getElementById('health-table');
  const rows = table ? table.querySelectorAll('tr') : [];
  rows.forEach(row => {{
    const cells = row.querySelectorAll('th, td');
    if (cells[index]) cells[index].style.display = checkbox.checked ? '' : 'none';
  }});
}}

function initHealthTable() {{
  const tbody = document.getElementById('health-body');
  healthData.forEach(d => {{
    const trend = d.cash_flow_trend;
    const trendClass = trend === 'Consistently Improving' ? 'trend-ci' :
                       trend === 'Recently Improving' ? 'trend-ri' :
                       trend === 'Recently Deteriorating' ? 'trend-rd' : 'trend-cd';
    const cash = '$' + (d.cash / 1e9).toFixed(2) + 'B';
    const ocf = d.ocf_current >= 0
      ? '+$' + (d.ocf_current / 1e9).toFixed(2) + 'B'
      : '-$' + (Math.abs(d.ocf_current) / 1e9).toFixed(2) + 'B';
    const ocfColor = d.ocf_current >= 0 ? '#4ade80' : '#f87171';
    const burnRate = d.burn_rate ? d.burn_rate.toFixed(3) : 'N/A';
    tbody.innerHTML += `
      <tr>
        <td class="ticker">${{d.ticker}}</td>
        <td>${{d.Security}}</td>
        <td><span class="sector-badge">${{d['GICS Sector']}}</span></td>
        <td>${{cash}}</td>
        <td style="color:${{ocfColor}};font-weight:600">${{ocf}}</td>
        <td>${{burnRate}}</td>
        <td><span class="badge ${{trendClass}}">${{trend}}</span></td>
      </tr>`;
  }});
}}

initLeaderboards();
initStripPlot();
initSectorFilter();
filterTable();
initHealthTable();
</script>
</body>
</html>"""

with open('earnings_quality_dashboard.html', 'w', encoding='utf-8') as f:
    f.write(html)

import os
os.startfile('earnings_quality_dashboard.html')
print("Dashboard saved and opened.")
```
